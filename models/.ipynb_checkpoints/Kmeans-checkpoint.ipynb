```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy import stats
from collections import Counter
```

First, the complete dataset is imported 

```python
df = pd.read_csv('first30.csv')
df.head()
```

As an example, the kmeans model is trained on the first 1000 playlists, so we construct the corresponding dataframe by cutting the original one.

```python
df = df[df['pid']<1000]
```

```python
df.head()
```

Now, the each playlist is split into a training set (the first 80%) and a validation set (the last 20%). The code below generates the training sets.

```python
def train_test_split(df, size=0.2):
    train = []
    test = []
    for i in range(len(df.pid.unique())):
        indices = np.array(df[df['pid']==i].index)
        train_indices = range(indices[0], indices[len(indices)//10*8])
        train += train_indices
    return df.iloc[train]
```

```python
train = train_test_split(df)
```

```python
train[train['pid']==1].head()
```

In order to effectively cluster data points, we first drop the text columns and variables without predictive power, then perform standarization on features.

```python
train_new = train.copy()
train_new = train.drop(columns = ['pos', 'artist_name', 'track_uri', 'artist_uri', 'track_name', 
                                  'album_uri', 'album_name'])
```

```python
scaler_list = [StandardScaler() for _ in range(len(train_new.columns))]
```

```python
for i, item in enumerate(list(train_new.columns)):
    if item!='pid':
        train_new[item] = scaler_list[i].fit_transform(np.array(train_new[item]).reshape(-1,1))
```

```python
train_new.head()
```

```python
train_new.describe()
```

Now we have a training set with desired columns and values with mean 0 and standard deviation 1, we can start clustering! A kmeans model with 50 clusters is used in this case.

```python
kmeans = KMeans(n_clusters=50, random_state=0).fit(train_new.values)
```

Two dataframes of unique songs are formed for later analysis. songs dataframe include all features for unique songs. songs_new include a standarized version of usable features on unique songs, which is used to pass to kmeans model for forecasting.

```python
songs = df.iloc[np.unique(df.track_uri, return_index=True)[1]]
songs_new = songs.copy()
songs_new = songs.drop(columns = ['pos', 'artist_name', 'track_uri', 'artist_uri', 'track_name', 
                                      'album_uri', 'album_name'])

for i, item in enumerate(songs_new.columns):
    if item !='pid':
        songs_new[item] = scaler_list[i].transform(np.array(songs_new[item]).reshape(-1,1))
```

To recommend songs. We first fit kmeans on all training songs to predict a list of cluster labels. Then, we take the majority label from the list for prediction. After running all unique songs in the dataset to obtain this group of songs, we recommend the first n songs to the user, where n is the length of the training set.

```python
def predict(train, songs_new):
    #first get the majority group in the playlist we want to evaluate
    groups = kmeans.predict(train)
    #g = stats.mode(groups)[0][0]
    #g=int(np.mean(groups))
    try:
        g = mode(groups)
    except:
        g = Counter(groups).most_common()[0][0]
#     print(g)
    # now get the unique songs and convert into features

    #now predict kmeans on every song and find out what is in the same group
    recommended = songs.iloc[kmeans.predict(songs_new.values) == g]
    #return the recommended tracks
    return recommended.track_uri
```

r-precision is used as a evaluation metric to evaluate the model's performance

```python
def eval_playlist(test_tracks, recommended_tracks): #r-precision
    test = set(test_tracks)
    pred = set(recommended_tracks)
    res = list(test & pred)
    return len(res) / len(test)
```

Now, we go through every playlist to generate recommendations and evaluate performance

```python
scores=[]
test = []
for i in train_new.pid.unique():
    train_tracks = train_new[train_new['pid']==i]
    
    length = len(df[df['pid']==i])
    train_uris = df[df['pid']==i].track_uri[:length//10*8]
    
    recommended_tracks = predict(train_tracks, songs_new)
    #remove already existing tracks
    recommend = [x for x in recommended_tracks if x not in train_uris][:length]
    
    test_tracks = df[df['pid']==i].track_uri[length//10*8:]
    #evaluate
    scores.append(eval_playlist(test_tracks, recommend))
```

```python
mean_score = np.mean(np.array(scores))
print('In the end, the kmeans model achieved a mean r-precision of ', mean_score)
```

```python
plt.subplots(figsize = (10,5))
plt.hist(np.array(scores), bins=30)
plt.title('Histogram of Score distribution')
plt.show()
```

From the distribution, we could see that most playlist (about half) still has a score of 0, which is expected because of the size of the playlists and the validation sets. Then, a majority of playlist scored around 8% r-precision, with best performing playlist generating r-precision of around 40%. Overall, kmeans is a good model that provides an improvement from the baseline model
